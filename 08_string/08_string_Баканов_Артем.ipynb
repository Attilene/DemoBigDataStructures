1.2 Напишите функцию `show_info`, которая для рецепта по его `id` создает строку (в смысле объекта python) с описанием следующего вида:

```
"Название"

1. Шаг 1.
2. Шаг 2.
----------
#тэг1 #тэг2
```

    
Данные для создания строки получите из файлов `recipes_sample_with_tags_ingredients.csv`, `steps_sample.xml` (__ЛР4__) и `tags_sample.csv` (__ЛР5__). 
Выведите созданную строку на экран.

````python
def show_info(ids: int, data: list()):
    s = '"{}"\n\n'.format(data[0].loc[data[0]["id"] == ids]["name"][0].capitalize())
    
    for recipe in data[1].find_all('recipe'):
        step_count = 1
        if recipe.find("id").text == str(ids):
            for step in recipe.find_all('step'):
                s += '{}. {} {}.\n'.format(str(step_count), step.text, str(step_count))
                step_count += 1
            break
            
    s += "---------------------\n"
    
    tags = data[2][data[2]['id'] == ids]
    for index, row in tags.iterrows():
        s += '#{} '.format(row["tag"])
    return s
````

3.2 Посчитайте кол-во уникальных слов в датасете (без учета регистра).

````python
# data4['date'] = pd.to_datetime(data4['date'])

buf = pd.DataFrame()
for key, value in data4.iteritems():
    if value.dtype == np.object:
        buf[key] = value
````

````python
unique_words = set()
for index, row in buf.iterrows():
    try:
        for elem in nltk.word_tokenize(row.iloc[0].lower()):
            unique_words.add(elem)
    except (TypeError, AttributeError):
        continue
````

3.3 Найдите 5 самых длинных (по количеству слов) отзывов в датасете и выведите их в порядке убывания длины.

````python
result = [[] for i in range(5)]
````

````python
for row in data4['review']:
    if type(row) == np.float:
        continue
    words = nltk.word_tokenize(row)
    if len(result[-1]) >= len(words):
        continue
    for index, elem in enumerate(result):
        if len(elem) >= len(words):
            continue
        result.insert(index, words)
        result = result[:5]
        break
````

3.4 Напишите функцию, которая для заданного предложения выводит информацию о частях речи слов, входящих в предложение в следующем виде:
```
PRP   VBD   DT      NNS     CC   VBD      NNS        RB   
 I  omitted the raspberries and added strawberries instead
``` 
Для определения части речи слова можно воспользоваться `nltk.pos_tag`.

Проверьте работоспособность функции на любом предложении из отзывов.


````python
nltk.download('averaged_perceptron_tagger')
````

````python
def part_of_speech(sentence):
    words, parts = '', ''
    for word, part in nltk.pos_tag(nltk.word_tokenize(sentence)):
        w = max(len(word), len(part))
        parts += f'{part:^{w}} '
        words += f'{word:^{w}} '
    print(parts)
    print(words)
````
